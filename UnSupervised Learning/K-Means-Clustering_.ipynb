```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Iris.csv')
df=df.drop(['Id','Species'],axis=1)
df
```

```python
from sklearn.cluster import KMeans
wcss_values=[]
for i in range(2,11):
  reg=KMeans(n_clusters=i)
  reg.fit(df)
  wcss_values.append(reg.inertia_)
```

```python
wcss_values
```

```python
plt.figure(figsize=(5,3))
plt.title('selecting no of best groups')
plt.xlabel('range of values')
plt.ylabel('wcss values')

plt.plot(np.arange(2,11),wcss_values,marker='*',color='r')
plt.show()
```

```python
reg=KMeans(n_clusters=5)
reg.fit(df)
```

```python
reg.cluster_centers_
```

```python
reg.labels_
```

```python
df.iloc[reg.labels_==3,2].shape
```

```python
plt.figure(figsize=(5,3))

plt.scatter(df.iloc[reg.labels_==0,0], df.iloc[reg.labels_==0,1],color='r')
plt.scatter(df.iloc[reg.labels_==1,0], df.iloc[reg.labels_==1,1],color='g')
plt.scatter(df.iloc[reg.labels_==2,0], df.iloc[reg.labels_==2,1],color='yellow')
plt.scatter(df.iloc[reg.labels_==3,0], df.iloc[reg.labels_==3,1])
plt.scatter(df.iloc[reg.labels_==4,0], df.iloc[reg.labels_==4,1])

plt.scatter(reg.cluster_centers_[:,0],reg.cluster_centers_[:,1],marker='*',color='black')
plt.show()
```
