```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
df = pd.DataFrame({"A1" : [0.40,0.22,0.35,0.26,0.08,0.45] ,
                   'B1' : [0.53,0.38,0.32,0.19,0.41,0.30]
                   })
df
```

```python
# giving the data to Hierarichal Algo
math_values=linkage(df)
math_values
```

```python
plt.figure(figsize=(5,3))
dendrogram(math_values)
plt.show()
```
