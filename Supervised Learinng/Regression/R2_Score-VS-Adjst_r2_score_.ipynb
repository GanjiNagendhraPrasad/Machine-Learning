#R2_SCORE VS ADJUSTED_R2_SCORE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/case_1.csv')
df
```

```python
X=df[['YearsExperience']]
y=df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
```

```python
print(f'{reg.coef_} * x + {reg.intercept_}')
```

```python
ans=reg.predict(X)
```

```python
from sklearn.metrics import r2_score
print(f'r2_score acc {r2_score(y,ans)}')
```

```python
X.shape[1]
```

```python
value = 1 - ( (1 - r2_score(y,ans)) * (len(df) - 1) ) / (len(df) - X.shape[1] - 1)
print(f'Accuracy by Adjusted r2_score was : {value}')
```

#case 2

```python
df=pd.read_csv('/content/case_2.csv')
df
```

```python
X = df[['YearsExperience','Height']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(f'{reg.coef_} * x + {reg.intercept_}')
```

```python
predictions = reg.predict(X)
```

```python
from sklearn.metrics import r2_score
print(f'Accuracy by r2_score was : {r2_score(y , predictions)}')
```

```python
X.shape[1]
```

```python
value = 1 - ( (1 - r2_score(y,predictions)) * (len(df) - 1) ) / (len(df) - X.shape[1] - 1)
print(f'Accuracy by Adjusted r2_score was : {value}')
```
