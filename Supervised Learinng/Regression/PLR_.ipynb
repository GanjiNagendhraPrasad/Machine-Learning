```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/Position_Salaries.csv')
df=df.drop(['Position'],axis=1)
df
```

```python
X=df[['Level']]
y=df['Salary']
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x=X,y=y,color='r',marker='*')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ans=reg.predict(X)
ans
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x=X,y=y,color='r',marker='*')
plt.plot(X,ans,color='b',marker='*')
plt.show()
```

#polynomial process

```python
plt.figure(figsize=(5,3))
plt.scatter(x=X,y=y,color='r',marker='*')
plt.show()
```

```python
X
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol=PolynomialFeatures()
pol.fit(X)
ans_pol=pol.transform(X)
ans_pol
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(ans_pol,y)
```

```python
ans=reg.predict(ans_pol)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x=X,y=y,color='r',marker='*')
plt.plot(X,ans,color='b',marker='*')
plt.show()
```
