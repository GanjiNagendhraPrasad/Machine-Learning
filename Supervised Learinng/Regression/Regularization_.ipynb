```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/kc_house_data.csv')
df=df.drop(['id','date'],axis=1)
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
x.columns
```

```python
reg.coef_
```

```python
t=pd.DataFrame()
t['columns']=x.columns
t['mvalues']=reg.coef_
```

```python
t
```

```python
from sklearn.linear_model import Ridge
pp=Ridge()
pp.fit(x,y)
```

```python
t['r-mvalues']=pp.coef_
```

```python
t
```

```python
from sklearn.linear_model import Lasso
ll=Lasso()
ll.fit(x,y)
```

```python
t['laso-mvalues']=ll.coef_
t
```
