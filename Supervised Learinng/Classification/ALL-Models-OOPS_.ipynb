```python
import warnings
warnings.filterwarnings('ignore')

import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.metrics import roc_auc_score,roc_curve

class ALL:
  try:
    def __init__(self,d):
      self.df=pd.read_csv(d)
      self.df=self.df.drop(['id','Unnamed: 32'],axis=1)
      self.df['diagnosis']=self.df['diagnosis'].map({'M':0,"B":1}).astype(int)
      self.x=self.df.iloc[:,1:]
      self.y=self.df.iloc[:,0]
      self.x_train,self.x_test,self.y_train,self.y_test=train_test_split(self.x,self.y,test_size=0.2,random_state=42)
  except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')


  def knn(self):
    try:
      self.knn_reg=KNeighborsClassifier(n_neighbors=5)
      self.knn_reg.fit(self.x_train,self.y_train)
      self.knn_pred=self.knn_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.knn_pred))
      print(accuracy_score(self.y_test,self.knn_pred))
      print(classification_report(self.y_test,self.knn_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')

  def nb(self):
    try:
      self.nb_reg=GaussianNB()
      self.nb_reg.fit(self.x_train,self.y_train)
      self.nb_pred=self.nb_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.nb_pred))
      print(accuracy_score(self.y_test,self.nb_pred))
      print(classification_report(self.y_test,self.nb_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def lr(self):
    try:
      self.lr_reg=LogisticRegression()
      self.lr_reg.fit(self.x_train,self.y_train)
      self.lr_pred=self.lr_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.lr_pred))
      print(accuracy_score(self.y_test,self.lr_pred))
      print(classification_report(self.y_test,self.lr_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')

  def dt(self):
    try:
      self.dt_reg=DecisionTreeClassifier(criterion='entropy')
      self.dt_reg.fit(self.x_train,self.y_train)
      self.dt_pred=self.dt_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.dt_pred))
      print(accuracy_score(self.y_test,self.dt_pred))
      print(classification_report(self.y_test,self.dt_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def rf(self):
    try:
      self.rf_reg=RandomForestClassifier(criterion='entropy',n_estimators=5)
      self.rf_reg.fit(self.x_train,self.y_train)
      self.rf_pred=self.rf_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.rf_pred))
      print(accuracy_score(self.y_test,self.rf_pred))
      print(classification_report(self.y_test,self.rf_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def adab(self):
    try:
      lr=LogisticRegression()
      self.adab_reg=AdaBoostClassifier(estimator=lr,n_estimators=5)
      self.adab_reg.fit(self.x_train,self.y_train)
      self.adab_pred=self.adab_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.adab_pred))
      print(accuracy_score(self.y_test,self.adab_pred))
      print(classification_report(self.y_test,self.adab_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def gb(self):
    try:
      self.gb_reg=GradientBoostingClassifier(n_estimators=5)
      self.gb_reg.fit(self.x_train,self.y_train)
      self.gb_pred=self.gb_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.gb_pred))
      print(accuracy_score(self.y_test,self.gb_pred))
      print(classification_report(self.y_test,self.gb_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def xgb(self):
    try:
      self.xgb_reg=XGBClassifier(n_estimators=5)
      self.xgb_reg.fit(self.x_train,self.y_train)
      self.xgb_pred=self.xgb_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.xgb_pred))
      print(accuracy_score(self.y_test,self.xgb_pred))
      print(classification_report(self.y_test,self.xgb_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')
  def svm(self):
    try:
      self.svm_reg=SVC(kernel='rbf')
      self.svm_reg.fit(self.x_train,self.y_train)
      self.svm_pred=self.svm_reg.predict(self.x_test)
      print(confusion_matrix(self.y_test,self.svm_pred))
      print(accuracy_score(self.y_test,self.svm_pred))
      print(classification_report(self.y_test,self.svm_pred))
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')

  def auc_roc(self):
    try:
      knn_fpr,knn_tpr,knn_th=roc_curve(self.y_test,self.knn_pred)
      nb_fpr,nb_tpr,nb_th=roc_curve(self.y_test,self.nb_pred)
      lr_fpr,lr_tpr,lr_th=roc_curve(self.y_test,self.lr_pred)
      dt_fpr,dt_tpr,dt_th=roc_curve(self.y_test,self.dt_pred)
      rf_fpr,rf_tpr,rf_th=roc_curve(self.y_test,self.rf_pred)
      adab_fpr,adab_tpr,adab_th=roc_curve(self.y_test,self.adab_pred)
      gb_fpr,gb_tpr,gb_th=roc_curve(self.y_test,self.gb_pred)
      xgb_fpr,xgb_tpr,xgb_th=roc_curve(self.y_test,self.xgb_pred)
      svm_fpr,svm_tpr,svm_th=roc_curve(self.y_test,self.svm_pred)

      plt.figure(figsize=(5,3))
      plt.plot([0, 1], [0, 1])
      plt.xlabel('fpr')
      plt.ylabel('tpr')
      plt.title('all models auc curves')

      plt.plot(knn_fpr,knn_tpr,label='knn')
      plt.plot(nb_fpr,nb_tpr,label='nb')
      plt.plot(lr_fpr,lr_tpr,label='lr')
      plt.plot(dt_fpr,dt_tpr,label='dt')
      plt.plot(rf_fpr,rf_tpr,label='rf')
      plt.plot(adab_fpr,adab_tpr,label='adab')
      plt.plot(gb_fpr,gb_tpr,label='gb')
      plt.plot(xgb_fpr,xgb_tpr,label='xgb')
      plt.plot(svm_fpr,svm_tpr,label='svm')
      plt.legend(loc = 0)
      plt.show()
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')

  def common(self):
    try:
      print('------knn------')
      self.knn()
      print('-----nb-----')
      self.nb()
      print('------lr-----')
      self.lr()
      print('----dt-----')
      self.dt()
      print('-----rf------')
      self.rf()
      print('------adab------')
      self.adab()
      print('-------gb------')
      self.gb()
      print('-----xgb------')
      self.xgb()
      print('------svm-----')
      self.svm()
      print('******** AUC_ROC ***********')
      self.auc_roc()
    except Exception as e:
      ms,ty,lin=sys.exc_info()
      print(f'{ms} {ty} {lin.tb_lineno}')

if __name__=="__main__":
  obj=ALL('/content/breast-cancer.csv')
  obj.common()
```

#cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

class CROSS:
  def __init__(self,d):
    self.df=pd.read_csv(d)
    self.df=self.df.drop(['id','Unnamed: 32'],axis=1)
    self.df['diagnosis']=self.df['diagnosis'].map({'M':0,"B":1}).astype(int)
    self.x=self.df.iloc[:,1:]
    self.y=self.df.iloc[:,0]

  def cross(self):
    rf=RandomForestClassifier(criterion='entropy',n_estimators=5)
    result=cross_val_score(rf,self.x,self.y,cv=10)
    print(result)
    print(len(result))
    print(f'test accuracy by cross validation {result.mean()}')

if __name__=="__main__":
  obj=CROSS('/content/breast-cancer.csv')
  obj.cross()
```

#Hyper Parameter Tuning Techniques

```python
df=pd.read_csv('/content/breast-cancer.csv')
df=df.drop(['id','Unnamed: 32'],axis=1)
df['diagnosis']=df['diagnosis'].map({'M':0,'B':1}).astype(int)
x=df.iloc[:,1:]
y=df.iloc[:,0]
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
parameters_list={'n_estimators' : [5,10,15,20,25] ,
                   'criterion' : ['gini' , 'entropy'] ,
                   'max_depth' : [2,3,4,5,6,7] ,
                   'class_weight':[None,"balanced", "balanced_subsample"],
                   }

rf=RandomForestClassifier()

reg=RandomizedSearchCV(estimator=rf,
                       param_distributions=parameters_list,
                       scoring='accuracy',
                       cv=10)
result=reg.fit(x,y)
```

```python
print(reg.best_params_)
```

```python
parameters_list={'n_estimators' : [5,10,15,20,25] ,
                   'criterion' : ['gini' , 'entropy'] ,
                   'max_depth' : [2,3,4,5,6,7] ,
                   'class_weight':[None,"balanced", "balanced_subsample"],
                   }
grid_reg=GridSearchCV(estimator=rf,
                       param_grid=parameters_list,
                       scoring='accuracy',
                       cv=10)
result=grid_reg.fit(x,y)
```
