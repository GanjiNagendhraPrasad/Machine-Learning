```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv('/content/Iris.csv')
df=df.drop(['Id'],axis=1)
df['Species']=df['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}).astype(int)
df.sample(10)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

#train performance

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
confusion_matrix(y_train,reg.predict(x_train))
```

```python
print(classification_report(y_train,reg.predict(x_train)))
```

```python
accuracy_score(y_train,reg.predict(x_train))
```

#test performance

```python
confusion_matrix(y_test,reg.predict(x_test))
```

```python
print(classification_report(y_test,reg.predict(x_test)))
```

```python
accuracy_score(y_test,reg.predict(x_test))
```

#checking with own data
