```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/breast-cancer.csv')
df
```

```python
df=df.drop(['id','Unnamed: 32'],axis=1)
df
```

```python
df.info()
```

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0}).astype(int)
```

```python
df
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train),len(y_train)
```

```python
len(x_test),len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
reg=KNeighborsClassifier(n_neighbors=5)  # default k value = 5
reg.fit(x_train,y_train)
```

```python
y_train.value_counts()
```

#train performance

```python
y_train.head(2)
```

```python
f=x_train.head(2)
f
```

```python
reg.predict(f)
```

```python
predictions=reg.predict(x_train)
```

```python
train_performance=pd.DataFrame()
train_performance=x_train.copy()
train_performance['actual_ans']=y_train
train_performance['predict_ans']=predictions
train_performance
```

```python
tp=0
tn=0
fp=0
fn=0

for i in train_performance.index:
  if train_performance['actual_ans'][i]==0 and train_performance['predict_ans'][i]==0:
    tn=tn+1
  elif  train_performance['actual_ans'][i]==1 and train_performance['predict_ans'][i]==1:
    tp=tp+1
  elif  train_performance['actual_ans'][i]==0 and train_performance['predict_ans'][i]==1:
    fp=fp+1
  else:
    fn=fn+1
print(f'True Positive : {tp}')
print(f'True negative : {tn}')
print(f'False Positive : {fp}')
print(f'False Negative : {fn}')
```

```python
print(f'Train acc {(tp+tn)/(tp+tn+fp+fn)}')
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_train,predictions)
```

```python
accuracy_score(y_train,predictions)
```

```python
print(classification_report(y_train,predictions))
```

#best k value

```python
odd=np.arange(3,51,2)
odd
```

```python
a=[]
for i in odd:
  reg=KNeighborsClassifier(n_neighbors=i)
  reg.fit(x_train,y_train)
  pred=reg.predict(x_test)
  a.append(accuracy_score(y_test,pred))
a
```

```python
odd[np.argmax(a)]
```

```python
plt.figure(figsize=(5,3))
plt.title('best k value')
plt.xlabel('odd numbers')
plt.ylabel('accuracy')
plt.plot(odd,a,color='r',marker='*')
plt.show()
```
